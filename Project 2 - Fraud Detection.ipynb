## Calculate Distance

```python
def haversine(lat1, lon1, lat2, lon2):
    # Konversi dari derajat ke radian
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    # Jarak dalam kilometer (radius bumi = 6371 km)
    distance = 6371 * c
    return distance



# Menghitung jarak untuk setiap transaksi
FraudData['distance'] = FraudData.apply(lambda row: haversine(row['lat'], row['long'], row['merch_lat'], row['merch_long']), axis=1)


# Misalkan df adalah DataFrame yang berisi kolom 'distance'
bins = [0, FraudData['distance'].quantile(0.25), FraudData['distance'].quantile(0.75), FraudData['distance'].max()]
labels = [1, 2, 3]

# Membuat kolom baru 'distance_category'
FraudData['distance_category'] = pd.cut(FraudData['distance'], bins=bins, labels=labels, include_lowest=True)
FraudData.head()



# Menghitung jarak untuk setiap transaksi
FraudTest['distance'] = FraudTest.apply(lambda row: haversine(row['lat'], row['long'], row['merch_lat'], row['merch_long']), axis=1)


# Misalkan df adalah DataFrame yang berisi kolom 'distance'
bins = [0, FraudTest['distance'].quantile(0.25), FraudTest['distance'].quantile(0.75), FraudTest['distance'].max()]
labels = [1, 2, 3]

# Membuat kolom baru 'distance_category'
FraudTest['distance_category'] = pd.cut(FraudTest['distance'], bins=bins, labels=labels, include_lowest=True)
FraudTest.head()
```

```python
industry_salary_map = [
    {
        'keywords': ['psychologist', 'nurse', 'therapist', 'physiotherapist', 'radiographer', 'pharmacist', 'doctor', 'medical'],
        'industry': 1,
        'us_salary_top': 150,
        'us_salary_down': 60
    },
    {
        'keywords': ['scientist', 'geologist', 'hydrologist', 'toxicologist', 'biochemist'],
        'industry': 2,
        'us_salary_top': 120,
        'us_salary_down': 55

    },
    {
        'keywords': ['engineer', 'developer', 'programmer', 'analyst', 'technician'],
        'industry': 3,
        'us_salary_top': 140,
        'us_salary_down': 70
    },
    {
        'keywords': ['teacher', 'lecturer', 'education officer'],
        'industry': 4,
        'us_salary_top': 85,
        'us_salary_down': 40
    },
    {
        'keywords': ['designer', 'editor', 'artist', 'presenter', 'director', 'copywriter'],
        'industry': 5,
        'us_salary_top': 100,
        'us_salary_down': 40
    },
    {
        'keywords': ['solicitor', 'lawyer', 'barrister', 'police', 'civil servant', 'attorney'],
        'industry': 6,
        'us_salary_top': 200,
        'us_salary_down': 55
    },
    {
        'keywords': ['accountant', 'trader', 'analyst', 'banker', 'manager', 'officer', 'cfo', 'cio'],
        'industry': 7,
        'us_salary_top': 250,
        'us_salary_down': 60
    },
    {
        'keywords': ['architect', 'surveyor', 'consultant', 'environmental'],
        'industry': 8,
        'us_salary_top': 110,
        'us_salary_down': 55
    },
    {
        'keywords': ['social worker', 'counsellor', 'probation', 'community'],
        'industry': 9,
        'us_salary_top': 80,
        'us_salary_down': 45
    },
    {
        'keywords': ['administrator', 'operations', 'logistics', 'programme'],
        'industry': 10,
        'us_salary_top': 120,
        'us_salary_down': 50
    }
]

def classify_job_row(job_title: str):
    job_title_lower = job_title.lower()
    for entry in industry_salary_map:
        if any(keyword in job_title_lower for keyword in entry['keywords']):
            return entry['industry'], entry['us_salary_top'],entry['us_salary_down']
    return 11 , 11 ,11

def enrich_with_salary_info(df: pd.DataFrame, title_col: str = 'Job Title') -> pd.DataFrame:
    df[['Industry', 'us_salary_top', 'us_salary_down']] = df[title_col].apply(
        lambda title: pd.Series(classify_job_row(str(title)))
    )
    return df




FraudData = enrich_with_salary_info(FraudData , "job")

FraudTest = enrich_with_salary_info(FraudTest , "job")

```
